```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_csv('/content/lotes-irpf-valor.csv', sep=';', decimal=',')

# Converter as colunas numéricas para valores numéricos
numeric_columns = df.columns[1:-1]  # Selecionar todas as colunas numéricas, exceto a primeira e a última
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.str.replace('.', '').str.replace(',', '.').astype(float))

# Plotar o gráfico
plt.figure(figsize=(12, 6))  # Definir o tamanho da figura

# Iterar sobre as colunas numéricas e plotar os dados
for column in numeric_columns:
    plt.plot(df['Mês'], df[column], marker='o', label=column)

plt.xlabel('Mês')
plt.ylabel('Valor')
plt.title('Restituição de Imposto de Renda')
plt.legend()
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para facilitar a leitura
plt.tight_layout()  # Ajustar o layout do gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_csv('/content/lotes-irpf-valor.csv', sep=';', decimal=',')

# Converter as colunas numéricas para valores numéricos
numeric_columns = df.columns[1:-1]
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.str.replace('.', '').str.replace(',', '.').astype(float))

# Filtrar apenas o mês de janeiro
df_janeiro = df[df['Mês'] == 'jan/2023']

# Plotar o gráfico de barras para o mês de janeiro
plt.figure(figsize=(12, 6))

for column in numeric_columns:
    value = df_janeiro[column].iloc[0]
    plt.bar(column, value, label=f'{column} ({value:.2f}%)')
    plt.annotate(f'{value:.2f}%', (column, value), ha='center', va='bottom')

plt.xlabel('Faixa de Renda')
plt.ylabel('Valor')
plt.title('Restituição de Imposto de Renda - Janeiro 2023')
plt.legend(fontsize='small')
plt.tight_layout()
plt.show()
```
